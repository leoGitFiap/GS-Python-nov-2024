```python
energ_solar = [150, 160, 155, 158, 145, 152, 147, 155, 160, 150] #(MW)
energ_eolica = [200, 210, 220, 215, 210, 205, 218, 212, 220, 205] #(MW)
energ_hidro = [500, 510, 505, 495, 480, 490, 500, 505, 510, 495] #(MW)

custo_solar = [80, 82, 79, 81, 80, 83, 82, 81, 84, 80] #($/MW)
custo_eolica = [60, 61, 62, 63, 60, 59, 62, 63, 61, 60] #($/MW)
custo_hidro = [50, 51, 50, 49, 48, 47, 50, 51, 50, 49] #($/MW)

def total_energ():

    "Função para calcular o total de energia produzida."

    total_solar = sum(energ_solar)
    total_eolica = sum(energ_eolica)
    total_hidro = sum(energ_hidro)

    print(f"solar = {total_solar}")
    print(f"eolica = {total_eolica}")
    print(f"hidrelétrica = {total_hidro}")

    return total_solar, total_eolica, total_hidro

res = total_energ()

print(f"O total de energia solar, eólica e hidrelétrica é: {res}, respectivamente.")

def custo():

    "Função para calcular o custo total de cada fonte"

    tot_custo_solar = 0
    tot_custo_eolica = 0
    tot_custo_hidro = 0

    for s in range(len(custo_solar)):
        
        tot_custo_solar += custo_solar[s] * energ_solar[s]

    for e in range(len(custo_eolica)):

        tot_custo_eolica += custo_eolica[e] * energ_eolica[e]

    for h in range(len(custo_hidro)):

        tot_custo_hidro += custo_hidro[h] * energ_hidro[h]

    return tot_custo_solar, tot_custo_eolica, tot_custo_hidro

res = custo()

print(f"O custo da energia solar, eólica e hidrelétrica é: {res}, respectivamente.")

def maior_menor():
    
    "Função que retorna o maior e o menor."

    maior_solar = max(energ_solar)
    menor_solar = min(energ_solar)
    
    maior_eolica = max(energ_eolica)
    menor_eolica = min(energ_eolica)

    maior_hidro = max(energ_hidro)
    menor_hidro = min(energ_hidro)

    return f"solar: {maior_solar, menor_solar} , eólica: {maior_eolica, menor_eolica} , hidrelétrica: {maior_hidro, menor_hidro}"

res = maior_menor()

print(f"Os maiores e os menores são, respectivamente: {res}")

def soma_prod_custo():

    "Função para somar valores de produção e de custo."

    tot_energ_solar = 0
    tot_energ_eolica = 0
    tot_energ_hidro = 0
    
    tot_custo_solar = 0
    tot_custo_eolica = 0
    tot_custo_hidro = 0

    for s in range(len(custo_solar)):
        
        tot_energ_solar += energ_solar[s]
        tot_custo_solar += custo_solar[s] * energ_solar[s]

    for e in range(len(custo_eolica)):

        tot_energ_eolica += energ_eolica[e]
        tot_custo_eolica += custo_eolica[e] * energ_eolica[e]

    for h in range(len(custo_hidro)):

        tot_energ_hidro += energ_hidro[h]
        tot_custo_hidro += custo_hidro[h] * energ_hidro[h]    

    return f"Produção Solar = {tot_energ_solar}, Custo Solar = {tot_custo_solar} ; Produção Eólica = {tot_energ_eolica}, Custo Eólico = {tot_custo_eolica} ; Produção Hidrelétrica = {tot_energ_hidro}, Custo Hidrelétrico = {tot_custo_hidro}"

res = soma_prod_custo()

print(f"A soma dos dados de produção e custo são: {res}")

def menor_custo_medio():

    "Função para verificar o menor custo médio."

    tot_custo_solar = 0
    tot_custo_eolica = 0
    tot_custo_hidro = 0

    for s in range(len(custo_solar)):
        
        tot_custo_solar += custo_solar[s] * energ_solar[s]

    for e in range(len(custo_eolica)):

        tot_custo_eolica += custo_eolica[e] * energ_eolica[e]
    
    for h in range(len(custo_hidro)):

        tot_custo_hidro += custo_hidro[h] * energ_hidro[h]

    media_s = tot_custo_solar / sum(energ_solar)
    media_e = tot_custo_eolica / sum(energ_eolica)
    media_h = tot_custo_hidro / sum(energ_hidro)

    if media_s < media_e and media_s < media_h:

        print(f"A menor média é: SOLAR = {media_s:.2f}")
    
    if media_e < media_s and media_e < media_h:

        print(f"A menor média é: EÓLICA = {media_e:.2f}")

    if media_h < media_s and media_h < media_e:

        print(f"A menor média é: HIDRELÉTRICA = {media_h:.2f}")

    return f"Custo Solar = {tot_custo_solar}, Média Solar = {media_s} ; Custo Eólico = {tot_custo_eolica}, Média eólica = {media_e} ; Custo Hidrelétrico = {tot_custo_hidro}, Média Hidrelétrica = {media_h}"

res = menor_custo_medio()

print(res)

def comp_custo_medio():

    def calculo_custo_medio(custo, energia):

        tot_custo = 0

        for i in range(len(custo)):

            tot_custo += custo[i] * energia[i]
        
        media = tot_custo / sum(energia)

        return media
    
    media_s = calculo_custo_medio(custo_solar, energ_solar)
    media_e = calculo_custo_medio(custo_eolica, energ_eolica)
    media_h = calculo_custo_medio(custo_hidro, energ_hidro)

    if media_s < media_e and media_s < media_h:

        print(f"A menor média é: SOLAR = {media_s:.2f}")
    
    if media_e < media_s and media_e < media_h:

        print(f"A menor média é: EÓLICA = {media_e:.2f}")

    if media_h < media_s and media_h < media_e:

        print(f"A menor média é: HIDRELÉTRICA = {media_h:.2f}")

    return (f"SOLAR = {media_s:.2f} ; EÓLICA = {media_e:.2f} ; HIDRELÉTRICA = {media_h:.2f}")

res = comp_custo_medio()

print(res)
```
